```python
SF12_3000m = pd.read_csv('data/split_data/3000m/data_3000m_SF_12.csv')
fig, axs = plt.subplots(1,3, figsize=(24,8))
axs[0].plot(SF12_3000m['RSSI'].rolling(window=2).mean(), label='SF12')
axs[0].axhline(y=SF12_3000m['RSSI'].describe().to_numpy()[1], color='blue', linestyle='--', alpha=0.8, label='SF12 mean')
axs[0].set_title('RSSI w/ mean')
axs[0].set_ylabel('dBm')
axs[0].legend()
axs[0].grid()

axs[1].plot(SF12_3000m['Packet RSSI'].rolling(window=2).mean(), label='SF12')
axs[1].axhline(y=SF12_3000m['Packet RSSI'].describe().to_numpy()[1], color='blue', linestyle='--', alpha=0.8, label='SF12 mean')
axs[1].set_title('Packet RSSI')
axs[1].set_xlabel('Received message number')
axs[1].set_ylabel('dBm')
axs[1].legend()
axs[1].grid()

axs[2].plot(SF12_3000m['SNR'].rolling(window=2).mean(), label='SF12')
axs[2].set_title('SNR')
axs[2].set_ylabel('dB')
axs[2].legend()
axs[2].grid()

plt.suptitle('3000m free space')
plt.tight_layout
plt.savefig('figs/3000m_free_space.png')
```
